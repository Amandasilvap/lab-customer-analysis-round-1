```python
import pandas as pd

df1 = pd.read_csv("files_for_lab/csv_files/file1.csv")
df1.shape
```

```python
df2 = pd.read_csv("files_for_lab/csv_files/file2.csv")
df2.shape
```

```python
df3 = pd.read_csv("files_for_lab/csv_files/file3.csv")
df3.shape
```

```python
df3.head(10)
```

```python
df1 = df1.rename(columns = {"Customer":"Id","ST":"State", "GENDER":"Gender","Customer Lifetime Value":"CLV","Number of Open Complaints":"N. Open Complaints"})
                        
df2 = df2.rename(columns = {"Customer":"Id","ST":"State", "GENDER":"Gender","Customer Lifetime Value":"CLV","Number of Open Complaints":"N. Open Complaints"})

df1.head()
```

```python
df3 = df3.rename(columns = {'Customer':'Id','Customer Lifetime Value':'CLV', 'Number of Open Complaints':'N. Open Complaints'})
                            
```

```python
df2 = df2[["Id","State","Gender","Education","CLV","Income","Monthly Premium Auto", "N. Open Complaints", "Policy Type","Vehicle Class","Total Claim Amount"]]
df3 = df3[["Id","State","Gender","Education","CLV","Income","Monthly Premium Auto", "N. Open Complaints", "Policy Type","Vehicle Class","Total Claim Amount"]]
```

```python
policies = pd.concat([df1,df2,df3], axis = 0)
policies
```

```python
policies.info()
```

```python
policies.dtypes
```

```python
policies = policies.drop("Education", axis = 1)
policies = policies.drop("N. Open Complaints", axis = 1)
```

```python
policies['CLV'] =  pd.to_numeric(policies['CLV'], errors='coerce')
policies['CLV'] =  policies['CLV']/100
policies.dtypes
```

```python
policies = policies.drop_duplicates()
policies.shape
```
